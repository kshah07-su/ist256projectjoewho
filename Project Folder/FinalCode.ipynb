```python
# Importing required packages
# Mainly worked on by Kushal, Drew and Drake helped in debugging
import pandas as pd
import requests 
import json
from datetime import date, timedelta

cols = ['temp']
list = [] # Creating an empty list

# Creating a start date and end date for which range of time we want the temperature for
start_date = date(2019, 4, 1) 
end_date = date(2019, 8, 30)

# Creating a loop to obtain temp for each date in the range created above
while start_date < end_date:
    delta = timedelta(days=1)
    start_date += delta # Increasing day by 1 so it collects data for the next day
    date = str(start_date) # Converting date to a string so it can be used in the url
    input_date = date+'T18:00:00'
    try: 
        url = 'https://api.darksky.net/forecast/8f09d4088f2426a3489e1e458f2b813c/43.078488,-76.165237,'+input_date # coordinates input for NBT Stadium
        result = requests.get(url)
        json_version = result.json()
        temperature = json_version['hourly']['data'][0]['temperature'] # get temperature at NBT Bank Stadium 
        list.append([temperature]) # adding temp data to a list
    except KeyError:
        print("Please use the correct format when entering a date.") # error for wrong format
    except requests.exceptions.RequestException as error:
        print("Error: can't connect to" , url) # error for no internet connection
        print(error)

weather = pd.DataFrame(list, columns=cols) # creating a data frame for the temperature obtained
```

```python
# Importing packages required
# Mainly worked on by Drew, Kushal and Drake helped in debugging

from datetime import date, timedelta

cols = ['date']
list2 = [] # Empty list

# Creating a start date and end date for which range of time we want the temperature for
start_date = date(2019, 4, 4)
end_date = date(2019, 8, 29)
delta = timedelta(days=1)
# Creating a loop to obtain each date in the range created above
while start_date <= end_date:
    a = start_date.strftime("%m/%d/%Y")
    list2.append(a) # adding the date to the list
    start_date += delta
```

```python
# Merging the date and temp together and adding it to the main csv file
# Mainly worked on by Drew, Kushal and Drake helped in debugging

date = pd.DataFrame(list2, columns=cols)
Weather = pd.merge(date,weather,left_index=True,right_index=True) # Merging the date and temp together

mets = pd.read_csv("SyracuseMets.csv")
Mets = pd.merge(mets,Weather,left_index=True,right_index=True) # Merging on the basis of date so it only adds the temp for gamedays
Mets.dropna() #Dropping N.A. values to make sure the regression works.
```

```python
# Importing to required packages
# Mainly worked on by Drake, Kushal and Drew helped in debugging
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.graphics.regressionplots import plot_ceres_residuals
from IPython.display import display, HTML

display(HTML("<h1>What Factors Influence Attendance at Syracuse Mets' Games?</h1>")) # Displaying the heading

# Displaying the instructions for users 
print ("Possible variables:")
print (" ")
print ("'num'(Game Number)")
print ("'month'(Number for month of the season with 4 being April and 8 being August)")
print ("'winpct'(Winning percentage of the team on the day of the game)")
print ("'temp'(Temperature on the day of the game at the field)")
print (" ")
print ("For accurate results choose 3 of the following 4 promotions the Mets used the past season:")
print ("'poster'\n'jersey'\n'bobblehead'\n'tshirt'")
print (" ")
print ("For accurate results choose 6 of the following 7 days of the week variables:")
print ("'sun'\n'mon'\n'tue'\n'wed'\n'thu'\n'fri'\n'sat'")
print (" ")
print ("For accurate results choose 4 of the following 5 months of the year variables:")
print ("'april'\n'may'\n'june'\n'july'\n'aug'")
print (" ")

# Using try and except to account for errors
# Mainly worked on by Kushal, Drew and Drake helped in debugging
try:
    variables = input("Please enter which variables you would like to use as independant variables (seprated by a space): ") # Asking the user for which independant variables
    indepvariable = variables.split(" ") # Splitting the input to create the variables as a list


    chart = Mets # Loading the csv
    x = chart[indepvariable]
    y = chart[['att']] # Dependant variable
    model = sm.OLS(y.astype(float),x.astype(float)) # Running the regression and converting the variables in the float
    fit = model.fit() #  Fitting the model
    print(fit.summary()) # Printing regression output
    
    # Creating a visualization that measures the influence
    sm.graphics.influence_plot(fit)
    plt.show()
    
    # Creating another visualization that measures each dependant variables' residual square
    
    fig = plt.figure(figsize=(16, 16))
    sm.graphics.plot_ccpr_grid(fit, fig=fig)
    plt.show()
    
except KeyError:
    print("Error: You did not follow directions when choosing variables!") # Error Message to be displayed
```
