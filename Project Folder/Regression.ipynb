```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.graphics.regressionplots import plot_ceres_residuals

print ("Possible variables:")
print (" ")
print ("'num'(Game Number)")
print ("'month'(Number for month of the season with 4 being April and 8 being August)")
print ("'winpct'(Winning percentage of the team on the day of the game)")
print ("'temp'(Temperature on the day of the game at the field)")
print (" ")
print ("Only can choose 6 of the following 7 days of the week variables:")
print ("'sun'")
print ("'mon'")
print ("'tue'")
print ("'wed'")
print ("'thu'")
print ("'fri'")
print ("'sat'")
print (" ")
print ("Only can choose 4 of the following 5 months of the year variables:")
print ("'april'")
print ("'may'")
print ("'june'")
print ("'july'")
print ("'aug'")

variables = input("Please enter which variables you would like to use as independant variables (seprated by a space): ")
indepvariable = variables.split(" ")

with open ('SyracuseMets.csv', 'r') as f:
    chart = pd.read_csv(f)

x = chart[indepvariable]
y = chart[['att']]
model = sm.OLS(y.astype(float),x.astype(float))
fit = model.fit()
print(fit.summary())
sm.graphics.influence_plot(fit)
plt.show()
plot_ceres_residuals(fit, indepvariable[0])
plt.show()
```
